```python
import os 
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Specify the directory containing the CSV files
directory_path = './results'

# Use glob to get all CSV files in the specified directory
csv_files = glob.glob(os.path.join(directory_path, "*.csv"))

# List to hold the DataFrames for each CSV file
dfs = []

# Iterate over the list of CSV files and read each file into a DataFrame
for file in csv_files:
    # print(file.split('_')[0])
    if file.split('_')[0] == './results/results':
        print(file)
        df = pd.read_csv(file)
        dfs.append((df, file))

# dfs
```

```python
for results_df, file in dfs:
    sns.set_theme(style="darkgrid")
    results_df = results_df.query('temperature == 0.001')
    # Grouped violinplot
    print(f'for file name {file}..')
    print(results_df.head())
    
    ax = sns.violinplot(x=results_df["model_name"], y=results_df["sim_diff"])
    # ax = sns.violinplot(x="day", y="total_bill", data=tips)

    # Set the title and labels
    ax.set_title(file_title_mapping[file])
    ax.set_xlabel('Model')
    ax.set_ylabel('Similarity Score Difference')

    plt.xticks(rotation=90)
    plt.show()
```
